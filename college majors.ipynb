```python
class jdf:
    
    def __init__(self,df):
        self.df = df
    
    def printcols(self):
        for c in self.df.columns:
            print c
            
    def compcols(self, odf):
        return self.df.columns == odf.columns
    
    def diffcols(self, odf):
        matches = self.compcols(odf)
        for i in range(len(matches)):
            if not matches[i]:
                print i + ": " + self.df.columns[i]
```

# What percent of each state has a bachelor's degree or higher

```python
def edu(df):
    ret = df.copy()[["Estimate; Total:","Estimate; Bachelor's degree or higher:"]]
    ret.columns = ["25_plus","bachelors"]
    ret["pct"] = ret["bachelors"] / ret["25_plus"]
    return ret

edu(ed_states_2010).sort_values(by="pct",ascending=False)
```

# Table 1. Percent of states' over 25 population with bach degrees
### 2010 versus 2015

```python
def edu_change(df_2010,df_2015):
    df10 = edu(df_2010)
    df15 = edu(df_2015)
    
    ret = df10.join(df15,lsuffix="_2010",rsuffix="_2015")
    ret["change"] = ret["pct_2015"] - ret["pct_2010"]
    
    return ret
    
edu_change(ed_states_2010,ed_states_2015).sort_values(by="change",ascending=False)
```

# Table 1. Percent of towns' over 25 population with bach degrees
### 2010 versus 2015

# Table 2: Male versus females with degrees, state level, 2015

```python
def male_female(df):
    ret = df.copy().set_index("Geography")
    ret = ret[["Total; Estimate; Total population 25 years and over with a Bachelor's degree or higher",
            "Males; Estimate; Total population 25 years and over with a Bachelor's degree or higher",
       "Females; Estimate; Total population 25 years and over with a Bachelor's degree or higher"]]
    ret.columns= ["total","males","females"]
    ret ["diff"] = (ret["females"] - ret["males"]) / ret["total"]
    return ret
    
male_female(states).sort_values(by="diff",ascending=False)
```

# Total degrees by degree category

```python
def by_degree(df):
    
    ret = df.copy().set_index("Geography")
    
    total_prefix = "Percent; Estimate; "
    
    cols = ["Total; Estimate; Total population 25 years and over with a Bachelor's degree or higher"]
    for d in degrees:
        cols.append(total_prefix + d)

    return ret[cols]


by_degree(states)
```

```python
def by_age(df, degree):
    ret = df.copy().set_index("Geography")
    
    # Sample format
    # Total; Estimate; DETAILED AGE - 40 to 64 years - Arts, Humanities and Others

    total_prefix = "Percent; Estimate; DETAILED AGE - "

    ages = {
        "25_39":"25 to 39 years",
        "40_64":"40 to 64 years",
        "65+":"65 years and over"
    }
    
    cols = []
    for k in ages:
        cols.append(total_prefix + ages[k] + " - " + degree)
        
    ret = ret[[total_prefix + ages["25_39"] + " - " + degree,
               total_prefix + ages["40_64"] + " - " + degree,
               total_prefix + ages["65+"] + " - " + degree,
              ]]
    ret.columns="25_39","40-64","65+"
    return ret

print degrees
#by_age(states,"Arts, Humanities and Others")
#by_age(states,"Science and Engineering")
```

```python
def ct_table():
    dfs = []
    for degree in degrees:
        df = by_age(states,degree)
        df["Field"] = degree
        dfs.append(df[df.index=="Connecticut"].set_index("Field"))
        
    ret = pd.concat(dfs)
    ret = ret[["65+","40-64","25_39"]]
    return ret#.transpose()
ct_table()#.sort_values(by="Education",ascending=False)
```

```python
def by_gender_age(topic):
    df = states[states["Geography"] == "Connecticut"]
    genders = ["Males","Females"]
    #cols = ["Percent Females; Estimate; DETAILED AGE - 65 years and over - Arts, Humanities and Others"]

    ages = {
        "25_39":"25 to 39 years",
        "40_64":"40 to 64 years",
        "65+":"65 years and over"
    }
    
    cols = []
    col_order = []
    for k in ages:
        col_order.append(ages[k])
        cols.append("Percent Females; Estimate; DETAILED AGE - " + ages[k] + " - " + topic)
        
    females = df[cols]

    females.columns = col_order
    females["group"] = "females"

    cols = []
    col_order = []
    for k in ages:
        col_order.append(ages[k])
        cols.append("Percent Males; Estimate; DETAILED AGE - " + ages[k] + " - " + topic)
        
    males = df[cols]
    males.columns = col_order

    males["group"] = "males"
    
    return pd.concat([males.set_index("group"),females.set_index("group")])[["65 years and over",
                                                                             "40 to 64 years",
                                                                             "25 to 39 years"
                                                                            ]]
    
```
